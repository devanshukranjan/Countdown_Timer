```python
import time
from tkinter import *
from tkinter import messagebox

running = True
run=True
#Creating Window
window=Tk()

#Giving Title to our window
window.title('Countdown Timer')
#Giving geometry to our window
window.geometry("400x300")
#To set background color
window.config(bg='#345')

hour=StringVar()
minute=StringVar()
second=StringVar()

def assign():
    global running
    global hourEntry
    global minuteEntry
    global secondEntry
    running=True
    hour.set("00")
    minute.set("00")
    second.set("00")

assign()
hourEntry = Entry(window, width=3, font=("Arial",18,""), textvariable=hour)
minuteEntry = Entry(window, width=3, font=("Arial",18,""), textvariable=minute)
secondEntry = Entry(window, width=3, font=("Arial",18,""), textvariable=second)
hourEntry.place(x=110,y=40)
minuteEntry.place(x=170,y=40)
secondEntry.place(x=230,y=40)

def submit():
    global running
    global run
    try:
        temp = int(hour.get())*3600 + int(minute.get())*60 + int(second.get())
    except:
        print("Please input the right value")
    while temp >-1:
        if running==False:
            break
        if run==False:
            break
        mins,secs = divmod(temp,60)
        hours=0
        if mins >60:
            hours, mins = divmod(mins, 60)
        
        hour.set("{0:2d}".format(hours))
        minute.set("{0:2d}".format(mins))
        second.set("{0:2d}".format(secs))
        
        window.update()
        time.sleep(1)
        if (temp == 0):
            messagebox.showinfo("Countdown Timer", "Time's up!! ")
            Reset()
        
        temp -= 1

def Start():
    global running
    global run
    run=True
    running=True
    start['state']='disabled'
    stop['state']='normal'
    reset['state']='disabled'
    pause['state']='normal'
    resume['state']='normal'
    submit()

def Stop():
    global running
    start['state']='disabled'
    stop['state']='disabled'
    reset['state']='normal'
    pause['state']='disabled'
    resume['state']='disabled'
    running = False

def Reset():
    assign()
    reset['state']='disabled'
    pause['state']='disabled'
    resume['state']='disabled'
    stop['state']='disabled'
    start['state']='normal'

def Pause():
    global run
    start['state']='disabled'
    stop['state']='normal'
    reset['state']='normal'
    pause['state']='disabled'
    resume['state']='normal'
    run=False

# Making buttons
start = Button(window, text='   START   ', bd='5', command= Start)
stop = Button(window, text='   STOP   ', bd='5',state='disabled', command=Stop)
reset = Button(window, text='   RESET   ',bd='5', state='disabled', command=Reset)
pause = Button(window, text='   PAUSE   ', bd='5',state='disabled', command=Pause)
resume = Button(window, text='   RESUME   ',bd='5', state='disabled', command=Start)
start.place(x = 156,y = 100)
stop.place(x = 115,y = 160)
reset.place(x = 195,y = 160)
pause.place(x = 109,y = 220)
resume.place(x = 190,y = 220)

window.mainloop()
```
